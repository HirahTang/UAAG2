```python
aa_list = ['Gly', 'Ala', 'Ser', 'Thr', 'Val', 'Leu', 'Ile', 'Asp', 'Glu',
       'Asn', 'Gln', 'Arg', 'Lys', 'His', 'Trp', 'Tyr', 'Phe', 'Cys', 'Pro',
       'Abu', 'Nva', 'Nle', 'Ahp', 'Aoc', 'TMe', 'hSM', 'tBu', 'Cpa', 'Cha',
       'Aib', 'MeG', 'MeA', 'MeB', 'MeF', '2Th', '3Th', 'YMe', '2Np', 'Bzt',
       'DAl']
```

```python
# convert cp2_df to another DataFrame, as the format of column 1: aa, column 2: pos, column 3: target, column 4: value
puma_reframe = pd.DataFrame(columns=['aa', 'pos', 'target', 'value'])
```

```python
for index, row in puma_df.iterrows():
    aa = row['aa']
    pos = row['pos']
    for target in aa_list:
        value = row[target]
        
        # check if the value is not NaN
        
    # convert target to all capital
    
    # create a new DataFrame with the same structure as cp2_reframe
        new_row = pd.DataFrame({'aa': [aa], 'pos': [pos], 'target': [target.upper()], 'value': [value]})
    
    # append the new DataFrame to cp2_reframe
        puma_reframe = pd.concat([puma_reframe, new_row], ignore_index=True)
```
